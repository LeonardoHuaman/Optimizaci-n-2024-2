### Gráfica de funciones

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return x**2

x = np.arange(-1, 1, 0.01)    # arreglo de posibles valores de x
y=np.array([f(t) for t in x])  # arreglo de las imagenes del arreglo anterior
```

```python
plt.plot(x, y)
```
