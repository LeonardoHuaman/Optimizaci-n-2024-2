# Funciones Unimodales

## Ejercicio 1

**Ejercicio 1:** Implementa las siguientes funciones.

a) $f(x)=x^4-3x^2+2x+1$

b) $g(x)=(x+1)^3+2$

c) $h(x)=e^{2x}+x^4$

Basandote en la gráfica de tales funciones. ¿Cuáles de ellas son unimodales?

```python
def f(x):
  return x**4 - 3*x**2 + 2*x + 1

def g(x):
  return (x+1)**3 + 2

def h(x):
  return np.exp(2*x) + x**4
```

```python
x = np.arange(-1,1,0.01)
y = np.array([f(t) for t in x])
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-5,2,0.01)
y = np.array([g(t) for t in x])
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-0.8,-0.4,0.01)
y = np.array([h(t) for t in x])
plt.plot(x, y)
plt.show()
```

### Por lo tanto:
a) NO ES UNIMODAL

B) NO ES UNIMODAL

C) SI ES UNIMODAL  

## Ejericio 2:

**Ejercicio 2:** Dada la función

$f(x)=x^4-3x^2+2x+1$

Utiliza el Método de Bisección para encontrar

a) Los puntos donde la función alcanza sus mímimos locales con una precisión de $10^{-5}$.

b) El punto donde la función alcanza su máximo local con una precisión de $10^{-5}$.

c) Los valores de los mínimos locales de la función y del máximo local de la función.

```python
def f_prima(x):
  return 4*x**3 - 6*x + 2

a = 0.5
b = 1.5
precision = 10**-5

while abs(b-a)/2>precision:
  x = (a+b)/2
  if f_prima(x)>0:
    b = x
  else:
    a = x

print(x)
```

## Ejercicio 3:

**Ejercicio 3:**

a) Implementa el método de bisección en una función. Que reciba como inputs:
* La derivada de la función objetivo
* Los límites del intervalo donde encontraremos el mínimo
* La precisión
y que de como output al mínimo de la función en dicho intervalo.

b) Utilizando la función de a), halla el mínimo de $h(x)=e^{2x}+x^4$ con una precisión de $10^{-5}$.

```python
def biseccion(f_prima, a, b, tol):
  # f_prima: derivada
  # a: extremo izquierdo
  # b: extremo derecho
  # tol: tolerancia
  while abs(b-a)/2>tol:
    x=(a+b)/2
    if f_prima(x)>0:
      b=x
    else:
      a=x
  return x
```

```python
biseccion(f_prima, -2, -1, 10**(-5))
```

```python
biseccion(f_prima, 0.5, 1.5, 10**(-5))
```

```python
def f_prima_opuesto(x):
  return -(4*x**3-6*x+2)
```

```python
biseccion(f_prima_opuesto, -1, 1, 10**(-5))
```

```python
def h_prima(x):
  return 2*np.exp(2*x)+4*x**3
```

```python
biseccion(h_prima, -1, 0, 10**(-5))
```
